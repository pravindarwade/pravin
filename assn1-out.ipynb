```python
n=int(input('enter the no of rows : '))
for i in range(1,n+1):
    for j in range(i):
        print('*',end='')
    print()  
for k in range(1,n+1):
    for l in range(1,n+1-k):
        print('*',end='')
    print()    
        
```
